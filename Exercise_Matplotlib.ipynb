# **Playing with Matplotlib**

## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Basic Graphs

```python
%matplotlib inline
# A simple linear plot
a = [2, 4, 6, 8]
b = [1, 3, 5, 7]
plt.plot(a, b)
plt.xlabel(xlabel='a')
plt.ylabel(ylabel='b')
plt.show()
```

```python
# There are various styles available in Matplotlib to style your plots!
print(plt.style.available)
```

```python
a = np.arange(-5, 5, 0.1)

# font = {'family': 'serif', 'color': 'black', 'weight': 'normal', 'size': 20}
# plt.figure(figsize=(5 ,3), dpi=200)   # dpi parameter is making the plot look large!!

plt.style.use("fivethirtyeight")                       # It nicely shows the plot replacing the 'font' dict and figsize
## format = '[color][marker][line]' like 'r.-' or 'b.--'
plt.plot(a, a**2, 'r--', label="y=x^2", linewidth=2)
plt.xlabel(xlabel="x")
plt.ylabel(ylabel="y")

#b = np.arange(-5, 5, 0.5)
plt.plot(a, np.exp(a), 'b-', label="y=exp(x)")
plt.xlabel(xlabel="x")
plt.ylabel(ylabel="y")

plt.title(label="Basic graphs")
plt.legend()
plt.show()
```

```python
## Sinusoidal functions
pi = np.pi
t = np.arange(-3*pi, 3*pi, 0.01*pi)

# font = {'family': 'monospace', 'color':'black', 'weight':'normal', 'size':20}
plt.figure(figsize=(7, 5), dpi=150)

plt.style.use("fivethirtyeight")
# Sin(x)
plt.plot(t, np.sin(t), 'g-', label='Sin(t)')
plt.xlabel(xlabel='t')
plt.ylabel(ylabel='y(t)')

# Cos(x)
plt.plot(t, np.cos(t), 'y-', label='Cos(t)')
plt.xlabel(xlabel='t')
plt.ylabel(ylabel='y(t)')

plt.title(label="Sinusoidal Functions")
plt.legend()
plt.show()
```

## Bar Chart

```python
labels = ['Cricket', 'Football', 'Hockey']
fans = [120, 60, 75]

plt.figure(figsize=(6, 4))
bars = plt.bar(labels, fans)

# bars[0].set_hatch('/')
# bars[1].set_hatch('+')
# bars[2].set_hatch('*')

patterns = ['/', '+', '*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

plt.show()
```



*   The increment in 2007-08 is due to economic crisis
*   Huge dip in gas prices in Japan and South Korea in year 1998 may be due to natural disasters, which frequently occure there



### FIFA Data

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head()
```

```python
fifa.shape
```

```python
fifa.isna().sum()
```

```python
popular_clubs = fifa.loc[fifa['Overall']>90, 'Club'].unique()
print(popular_clubs)
```

```python
mean_net_worth = []
for club in popular_clubs:
    values = list(fifa.loc[(fifa['Overall']>90) & (fifa['Club']==club), 'Value'])
    mean_net_worth.append(np.mean(np.array([float(x.strip('').strip('M')) for x in values])))
```

```python
mean_net_worth = sorted(mean_net_worth, reverse=True)
print(mean_net_worth)
```

```python
# plt.style.use("fivethirtyeight")
plt.figure(figsize=(10, 5))
plt.barh(popular_clubs, mean_net_worth)
plt.title("Mean net worth of players of different clubs (in )")      # Horizontal bar chart
plt.ylabel(ylabel='Football club')
plt.xlabel(xlabel="Mean_net_worth ()")
plt.show()
```

#### Histogram

```python
plt.style.use("_classic_test_patch")
plt.figure(figsize=(6, 5))
plt.hist(fifa['Overall'], color='#3e3960')
plt.xlabel(xlabel='Skill level')
plt.ylabel(ylabel="No. of Players")
plt.title("Distribution of players' skills in FIFA", fontdict={'family': 'serif', 'weight': 'bold', 'size': 15})
plt.show()
```

#### Pie Chart

```python
# Number of lefty and righty players
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

plt.style.use("seaborn")
plt.pie(x=[left, right], labels=['Left foot', 'Right foot'], colors=['#de3121', '#e6b517'], autopct='%0.2f %%', shadow=True)
plt.title("Foot preference of FIFA players", fontdict={'family': 'serif', 'weight': 'normal', 'size':15})
plt.show()
```

```python
# Removing lbs from weight
fifa['Weight'] = [int(x.strip('lbs')) if type(x) == str else x for x in fifa['Weight']]
fifa['Weight'][0]
```

```python

# Grouping by player weights (lbs)
light = fifa.loc[fifa['Weight']<125, 'Weight'].count()
light_medium = fifa.loc[(fifa['Weight']>=125) & (fifa['Weight']<150), 'Weight'].count()
medium = fifa.loc[(fifa['Weight']>=150) & (fifa['Weight']<175), 'Weight'].count()
medium_heavy = fifa.loc[(fifa['Weight']>=175) & (fifa['Weight']<200), 'Weight'].count()
heavy = fifa.loc[fifa['Weight']>200, 'Weight'].count()
light
```

```python
# Make pie chart for above groups
weight_groups = np.array([light, light_medium, medium, medium_heavy, heavy])
labels = ['Under125', '125-150', '150-175', '175-200', 'Over200']
explode = np.array([0.5, 0.2, 0, 0, 0.5])

plt.figure(figsize=(7, 7))
plt.style.use('fivethirtyeight')
plt.pie(x=weight_groups, labels=labels, autopct='%0.2f %%', pctdistance=0.8, explode=explode, shadow=True)
plt.title("FIFA player groups by weights (lbs)")
plt.show()
```

#### BoxPlot

```python
print(plt.style.available)
```

```python
# Some club comparison using boxplot
barca = fifa.loc[fifa['Club']=='FC Barcelona', 'Overall']
madrid = fifa.loc[fifa['Club']=='Real Madrid', 'Overall']
juventus = fifa.loc[fifa['Club']=='Juventus', 'Overall']
cambridge = fifa.loc[fifa['Club']=='Cambridge United', 'Overall']

labels = ['FC Barcelona', 'Real Madrid', 'Juventus', 'Cambridge United']
plt.figure(figsize=(7, 7))
plt.style.use("grayscale")
boxes = plt.boxplot([barca, madrid, juventus, cambridge], labels=labels, patch_artist=True, medianprops={'linewidth': 2})

for box in boxes['boxes']:
    box.set(color='#3c3b52', linewidth=2)   # edge color and width
    box.set(facecolor='#f0814d')            # fill color in the boxes

plt.title("Soccer club comparison", fontdict={'family': 'serif', 'weight': 'normal', 'size':15})           
plt.show()
```

The first three clubs are, as many football fans know, are world class and that can be seen from above boxplot. There is not any big difference between their overall goals median value in FIFA. While a local football club in England has a big difference in 'Overall' goals in FIFA as compared to the other two.
